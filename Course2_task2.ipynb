# Dealing with non-numeric data (One-Hot Encoding)

Categorical variables may be used directly as predictor or predicted variables in a multiple regression model as long as they've been converted to binary values. In order to pre-process the sales data as needed we first need to convert all object variables to binary features that contain ‘0’ and ‘1’ classes. Fortunately, pandas has a method for creating these 'Dummy Variables' as follows:

```python
# Let's start by checking our datatypes:
credit.dtypes
```

We can see we have three variables that are listed as being the object data type; we need to covert these to 'Dummy variables' so they can be used in this multiple regression problem.

```python
# Use the get_dummies() function within pandas to convert non-numeric columns to a series of binary numeric 'Dummy' columns
credit1 = pd.get_dummies(credit)
# Print credit
credit1
```

The categorical variables SEX and EDUCATION are now converted to dummy variables.

```python
#Double-check dtypes at this point to make sure everything is now numeric as needed.
credit1.dtypes
```

Note that the dummy variables have uint8 as data type. A uint8 data type contains all whole numbers from 0 to 255. As with all unsigned numbers, the values must be non-negative.

# Histograms

Histograms are very useful for understanding how the data is distributed in any given single (univariate) variable.

First, import the matplotlib library; it contains most of the major visualizations we'll be using. (Note: Another popular Python library is called Seaborn; you'll often see either it or matplotlib used in Data Science, and the two are similar.)

```python
import pandas as pd
import matplotlib.pyplot as plt # visualize data graphically
```

Next, lets make sure we know the column names in our dataframe so we can begin to build plots for each, where needed. We can do this with the index object in Pandas as follows:

```python
header = credit.dtypes.index
print(header)
```

Finally, lets build our first histogram using the .hist method of matplotlib as follows; this example plots the 'LIMIT_BAL' variable:

```python
plt.hist(credit['LIMIT_BAL'])
plt.show()
```

```python
plt.hist(credit['AGE'],bins=6)
plt.show()
```

Note that from Pandas-Profiling report, we can see individual BILL_AMTx distribution. I am going to plot all six BILL_AMTx in one plot. 
Because they have big spreads, and the histograms are squeezed in the middle, I am going to set the limits for the x-axis.

```python
plt.hist(credit['BILL_AMT1'], bins=16, label= 'Bill_Amt1',alpha = 0.5)
plt.hist(credit['BILL_AMT2'], bins=16, label= 'Bill_Amt2',alpha = 0.5)
plt.hist(credit['BILL_AMT3'], bins=16, label= 'Bill_Amt3',alpha = 0.5)
plt.hist(credit['BILL_AMT4'], bins=16, label= 'Bill_Amt4',alpha = 0.5)
plt.hist(credit['BILL_AMT5'], bins=16, label= 'Bill_Amt5',alpha = 0.5)
plt.hist(credit['BILL_AMT6'], bins=16, label= 'Bill_Amt6',alpha = 0.5)
plt.gca().set(title='Frequency Histogram of Bill Amounts', ylabel='Frequency')
plt.xlim(-100000,550000)
plt.legend()
plt.show()
```

Well, the distributions for the 6 bills do not seem too different.
But let's normalize it by setting density=True and stacked=True, to make total area in the histogram to be 1.

```python
# Normalize
kwargs = dict(alpha=0.5, bins=16, density=True, stacked=True)

# Plot
plt.hist(credit['BILL_AMT1'], **kwargs, label= 'Bill_Amt1')
plt.hist(credit['BILL_AMT2'], **kwargs, label= 'Bill_Amt2')
plt.hist(credit['BILL_AMT3'], **kwargs, label= 'Bill_Amt3')
plt.hist(credit['BILL_AMT4'], **kwargs, label= 'Bill_Amt4')
plt.hist(credit['BILL_AMT5'], **kwargs, label= 'Bill_Amt5')
plt.hist(credit['BILL_AMT6'], **kwargs, label= 'Bill_Amt6')
plt.gca().set(title='Probability Histogram of Bill Amounts', ylabel='Probability')
plt.xlim(-100000,500000)
plt.legend()
plt.show()
```

Because the distributions are similar for the BILL_AMTx, no dominating Month, so the probability histograms do not provide any more information that the original frequency plots.

# Box Plots

Box plots, sometimes called Box and Whisker Plots, are very useful for quickly visualizing where the bulk of the data is in any given feature ad are also very useful for identifying feature with outliers or extreme variances in the observations. As you already know understanding what features might contain outliers is very important in Data Science and very important for Machine Learning. Lets look at the different components that make up a good box plot.

The lines that extend from either side of the 'box' in the middle of the plot can demonstrate the existence of outlying data points in the feature if point go beyond the ends of the lines (whiskers), but this doesn't necessarily mean there is an issue.

Create box plots in matlotlib. We'll start again a having a look at our column names:

```python
header = credit.dtypes.index
print(header)
```

```python
# build a plot plot with BILL_AMT1
A = credit['BILL_AMT1']
plt.boxplot(A,0,'gD')
plt.show()
```

# Correlation

Whether addressing collinearity or checking for relationships between features for selection or engineering, calculating correlation is a vital part of all Data Science projects:

```python
# building a basic correlation coefficient matrix using pandas:
# Strength of correlations:
#   Very strong relationship  (|r| >= 0.8)
#   Strong relationship       (|r| >= 0.6)
#   Moderate relationship     (|r| >= 0.4)
#   Weak relationship         (|r| >= 0.2)
#   Very weak relationship    (0< |r|< 0.2)

corrMat = credit1.corr()
print(corrMat)
```

LIMIT_BAL has

1) very weak positive correlation to age

2) very weak negative correlation to MARRIAGE

3) weak negative correlation to PAY_x (=> LIMIT_BAL increases as pay in full or no consumption, LIMIT_BAL decreases as paymentdelinquent as month increase)

4) weak positive correlation to BILL_AMTx and PAY_AMTx where x = 1,..,6

5) very weak positive correlation PAY_AMT1 and PAY_AMT2, weak positive correlation to PAY_AMTx where x = 3,..,6

6) No linear correlation with gender

7) weak positive correlation to EDUCATION graduate_school

8) weak negative correlation to default_payment_next_month_default

default_payment_next_month_default has

1) no linear correlation with age, gender, marital status, education,BILL_AMTx, PAY_AMTx where x = 1,..,6

2) very weak negative correlation to LIMIT_BAL

3) weak positive correlation to PAY_x, where x = 0, 2, 3, 4, 5, 6. Note that the strength of the linear correlations decrease as x increases.

```python
# Create Pearson correlation matrix visualization, and limit to 2 decimal places:
corrMat.style.background_gradient(cmap='coolwarm').set_precision(2)
```

# Covariance

Covariance is often used to gauge the linear degree of change between two variables. This will be very important when studying the impact various features might have on default rates

```python
# To calculate covariance in Pandas enter the following
covMat = credit.cov()
print(covMat)
```

The concern about using COVARIANCE matrix is that the scale (or range) of the variables may cause some variables to dominate, in this case will be LIMIT_BAL, BILL_AMTx and PAY_AMTx. 

```python
# Create Pearson correlation matrix visualization, and limit to 2 decimal places:
covMat.style.background_gradient(cmap='coolwarm').set_precision(2)
```
